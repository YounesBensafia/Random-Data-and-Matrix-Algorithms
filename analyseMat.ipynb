```python
import csv
import matplotlib.pyplot as plt

size = []
time = []

with open('./complexityMul.csv', 'r') as file:
    reader = csv.reader(file)
    next(reader)

    for row in reader:
        size.append(row[0])
        time.append(float(row[1]))
```

```python
import numpy as np
def moving_average(data, window_size):
    return np.convolve(data, np.ones(window_size)/window_size, mode='valid')

linwidth = 5
window_size = 100 

x = size[window_size-1:]
y_smooth = moving_average(time, window_size)

plt.figure(figsize=(25, 10))
plt.plot(x, y_smooth, linewidth=linwidth)

plt.legend(['SIZE', 'TIME'])
plt.show()
```
