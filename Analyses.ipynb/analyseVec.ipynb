```python
import csv
import matplotlib.pyplot as plt

valeur = []
seq = []
dicho = []

with open('./Complexity_csvS/complexity.csv', 'r') as file:
    reader = csv.reader(file)
    
    next(reader)
    
    for row in reader:
        valeur.append(float(row[0]))
        seq.append(float(row[1]))
        dicho.append(float(row[2]))
```

```python
import numpy as np
def moving_average(data, window_size):
    return np.convolve(data, np.ones(window_size)/window_size, mode='valid')

linwidth = 1
window_size = 800 

x = valeur[window_size-1:]
y_smooth = moving_average(seq, window_size)
z_smooth = moving_average(dicho, window_size)

plt.figure(figsize=(10, 10))
plt.plot(x, y_smooth, linewidth=linwidth)
plt.plot(x, z_smooth, linewidth=linwidth)

plt.legend(['SEQ', 'DICHO'])
plt.show()
```
