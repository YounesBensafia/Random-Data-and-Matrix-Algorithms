```python
import csv
import numpy as np

import matplotlib.pyplot as plt

power = []
time = []

with open('./Complexity_csvS/complexityPow.csv', 'r') as file:
    reader = csv.reader(file)
    next(reader)
    for row in reader:
        power.append(row[0])
        time.append(float(row[1]))
```

```python
def moving_average(data, window_size):
    return np.convolve(data, np.ones(window_size)/window_size, mode='valid')

linwidth = 3
window_size = 10 
x = power[window_size-1:]
y_smooth = moving_average(time, window_size)
plt.figure(figsize=(30, 5))
plt.plot(x, y_smooth, linewidth=linwidth)

plt.legend(['POWER', 'TIME'])
plt.show()
```
